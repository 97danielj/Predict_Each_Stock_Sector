```python
from pykiwoom.kiwoom import *
import time
import pandas as pd


# TR 요청 (연속조회) - 단일조회
dfs = []
df = kiwoom.block_request("opt10081",
                          종목코드="005930",
                          기준일자="20200424",
                          수정주가구분=1,
                          output="주식일봉차트조회",
                          next=0)
print(df.head())
dfs.append(df)
```

```python
print(type(dfs[0]))
```

```python
print(df.shape)
print('---------------')
#600\ 붇이기 
while kiwoom.tr_remained: #데이터 남았으면 연속조회
    df = kiwoom.block_request("opt10081",
                              종목코드="005930",
                              기준일자="20200424",
                              수정주가구분=1,
                              output="주식일봉차트조회",
                              next=2) 
    dfs.append(df)
    time.sleep(1)

df = pd.concat(dfs) # => 데이터프레임 교환

print(df)
```
